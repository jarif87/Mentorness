# 1. Team Performance Analysis:
- Explore team-wise performance metrics.
- Analyze runs scored, wickets taken, and batting/bowling styles.
- Identify top-performing teams and their strengths.

```python
team_stats = df.groupby('team').agg({
    'runs': 'sum',
    'wkts': 'sum',
    'runs_per_ball': 'mean',
    'econ': 'mean',
    'wicketball_prob': 'mean'
}).reset_index()


top_teams = team_stats.nlargest(5, 'runs')  # Example: top 5 teams by total runs scored


print("Top Performing Teams:")
print(top_teams)


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_teams['team'], top_teams['runs'], color='skyblue')
plt.xlabel('Team')
plt.ylabel('Total Runs Scored')
plt.title('Top Performing Teams by Total Runs Scored')
plt.xticks(rotation=45)
plt.show()
```

# 2. Player Performance Analysis:
- Evaluate individual player statistics for both batting and bowling.
- Identify leading run-scorers and wicket-takers.
- Assess the impact of players on their team's performance.

```python
batting_stats = df.groupby('player').agg({
    'runs': 'sum',
    '4s': 'sum',
    '6s': 'sum',
    'mins': 'sum',
    'not_out': 'sum'
}).reset_index()
```

```python
batting_stats['centuries'] = batting_stats['runs'] // 100
batting_stats['half_centuries'] = batting_stats['runs'] // 50 - batting_stats['centuries']
batting_stats['average'] = batting_stats['runs'] / (batting_stats['not_out'] + 1)
batting_stats['strike_rate'] = (batting_stats['runs'] / batting_stats['mins']) * 100
```

```python
leading_run_scorers = batting_stats.nlargest(5, 'runs')

print("Leading Run-Scorers:")
print(leading_run_scorers)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(leading_run_scorers['player'], leading_run_scorers['runs'], color='orange')
plt.xlabel('Player')
plt.ylabel('Total Runs Scored')
plt.title('Leading Run-Scorers')
plt.xticks(rotation=45)
plt.show()
```

# 4. Temporal Analysis:
- Study performance trends over time, considering start dates and overs played.
- Identify any temporal patterns or changes in team and player performance.

```python

df['start_date'] = pd.to_datetime(df['start_date'])


df['year_month'] = df['start_date'].dt.to_period('M')

temporal_stats = df.groupby('year_month').agg({
    'runs': 'mean',
    'wkts': 'mean',
    'econ': 'mean',
    'overs': 'sum'  
})

```

```python
temporal_stats
```

```python

plt.figure(figsize=(12, 6))
plt.plot(temporal_stats.index.astype(str), temporal_stats['runs'], label='Average Runs Conceded')
plt.plot(temporal_stats.index.astype(str), temporal_stats['wkts'], label='Average Wickets Taken')
plt.plot(temporal_stats.index.astype(str), temporal_stats['econ'], label='Average Economy Rate')
plt.xlabel('Year-Month')
plt.ylabel('Average Performance')
plt.title('Performance Trends Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
